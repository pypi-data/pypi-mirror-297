```python
import toyplot
```

```python
canvas = toyplot.Canvas(style={"background-color":"ivory"}, width=200, height=200)
table=canvas.table(rows=2, columns=2)
table.cells.grid.hlines[...] = "single"
table.cells.grid.vlines[...] = "single"
```

```python
canvas = toyplot.Canvas(style={"background-color":"ivory"}, width=200, height=200)
table=canvas.table(rows=2, columns=2, margin=30)
table.cells.grid.hlines[...] = "single"
table.cells.grid.vlines[...] = "single"
```

```python
canvas = toyplot.Canvas(style={"background-color":"ivory"}, width=200, height=200)
table=canvas.table(rows=2, columns=2, margin=(30, 60))
table.cells.grid.hlines[...] = "single"
table.cells.grid.vlines[...] = "single"
```
