```python
import toyplot
canvas = toyplot.Canvas(style={"background-color":"lightgray"}, hyperlink="#canvas")
axes = canvas.cartesian(hyperlink="#axes")
axes.text(0, 0, "Why not <a href='#text'>click me?</a>");
```

```python
canvas = toyplot.Canvas()
axes = canvas.cartesian()
axes.bars([1, 2, 3], hyperlink=["#1", "#2", "#3"]);
```

```python
canvas = toyplot.Canvas()
axes = canvas.cartesian()
axes.bars([[1, 2], [3, 4], [5, 6]], hyperlink=["#1", "#2", "#3"], baseline=None);
```

```python
canvas = toyplot.Canvas(width=300)
canvas.text(
    150,
    150,
    "<a href='http://www.sandia.gov' target='_blank'>sandia.gov</a>",
    style={
        "font-size": "24px",
        "text-decoration-line":"none",
        "fill": "steelblue",
    },
);
```
