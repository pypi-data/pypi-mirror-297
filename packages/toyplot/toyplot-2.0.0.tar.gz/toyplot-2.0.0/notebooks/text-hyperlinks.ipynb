```python
import toyplot.font
import toyplot.text
```

```python
style = {}
fonts = toyplot.font.ReportlabLibrary()
layout = toyplot.text.layout("<a href='#'>Foo</a>", style, fonts)

toyplot.text.dump(layout)
```

```python
canvas = toyplot.Canvas()
axes = canvas.cartesian()
axes.x.ticks.locator = toyplot.locator.Explicit(labels=["<a href='http://www.sandia.gov'>foo</a>", "bar", "baz"])
axes.text(0, 0, "Hello from <a href='#foo'>foo</a> and <a href='#bar'>bar</a>!")
```
