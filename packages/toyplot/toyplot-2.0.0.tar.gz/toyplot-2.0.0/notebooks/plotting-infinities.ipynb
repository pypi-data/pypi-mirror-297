```python
import toyplot.data, toyplot.locator, toyplot.pdf
import numpy
numpy.random.seed(1234)
```

```python
data = toyplot.data.Table()
data["y"] = numpy.concatenate((
        1e+5 + numpy.power(numpy.linspace(0, 10, 10), 6),
        [1e+5 + numpy.power(10, 6)],
        1e+5 + numpy.power(numpy.linspace(-10, 10, 20), 6),
#        1e+5 + numpy.power(numpy.linspace(-10, 0, 10), 6),
        ))
data["y"] *= numpy.random.uniform(0.9, 1/0.9, size=len(data))
data["y"][8:13] = numpy.inf
data["y"][10:12] = numpy.nan
data["y"][29:31] = numpy.inf
data["x"] = numpy.arange(len(data))

minimum = numpy.nanmin(data["y"])
maximum = numpy.nanmax(data["y"][~numpy.isinf(data["y"])])

data = data.columns(["x", "y"])
```

```python
canvas, table = toyplot.table(data, width="2.5in", height="8in")
table.column(0).column_offset = -25
table.column(1).format = toyplot.format.FloatFormatter("{:.0f}")
```

```python
toyplot.pdf.render(canvas, "plotting-infinities-data.pdf")
```

```python
canvas, axes, plot = toyplot.plot(data["x"], data["y"], marker="o", label="Case 1", width="6in", height="4in")
#axes.y.ticks.locator = toyplot.locator.Explicit([minimum, maximum])
```

```python
toyplot.pdf.render(canvas, "plotting-infinities-1.pdf")
```

```python
canvas, axes, plot = toyplot.plot(data["x"], data["y"], marker="o", ymax=400000, label="Case 2", width="6in", height="4in")
#axes.y.ticks.locator = toyplot.locator.Explicit([minimum, maximum])
```

```python
toyplot.pdf.render(canvas, "plotting-infinities-2.pdf")
```
