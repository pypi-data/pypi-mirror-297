```python
import sys
sys.path.append("../features/steps")

import numpy

import testing
import toyplot.color
import toyplot.layout

numpy.random.seed(1234)
```

```python
G = testing.prufer_tree(numpy.random.choice(4, 12))
```

```python
colormap = toyplot.color.LinearMap(toyplot.color.Palette(["white", "yellow", "red"]))
```

```python
canvas = toyplot.Canvas(width=1000, height=1000)

for i in range(4):  
    layout = toyplot.layout.graph(G, layout=toyplot.layout.Random(seed=1234+i))
    
    axes = canvas.cartesian(grid=(2, 2, i), show=False)
    axes.aspect = "fit-range"
    mark = axes.graph(
        G,
        olayout=layout,
        vcolor=colormap,
        vstyle={"stroke":"black"},
        vsize=20,
        ecolor="black",
        eopacity=0.2,
    )
```
