```python
import numpy
import toyplot
```

```python
hlines = [1, 2, 4, 8, 11, 13]

locations = numpy.arange(1, 14)
labels = ["Complicated task {}".format(number) for number in locations]
labels[-1] = ""
```

```python
canvas = toyplot.Canvas(width=800, height=400)
axes = canvas.cartesian(margin=(50, 50, 50, 200))

# So the spine touches the hlines
axes.padding = 0
# Create long ticks that extend
# from the spine away from the domain
axes.y.ticks.near = 125 
axes.y.ticks.far = 0
axes.y.ticks.show = True

# Horizontal labels (automatically right-justified),
# offset so they don't cover the ticks, for a subset
# of the hlines
axes.y.ticks.labels.angle = -90
axes.y.ticks.labels.style = {"baseline-shift": "100%", "-toyplot-anchor-shift":"-10px"}
axes.y.ticks.locator = toyplot.locator.Explicit(locations=locations, labels=labels)

# Hide the ticks that don't have matching hlines
for index, location in enumerate(locations):
    if location not in hlines:
        axes.y.ticks.tick(index).style = {"stroke":"none"}

axes.hlines(hlines, style={"stroke-dasharray":"1,3"});
```
