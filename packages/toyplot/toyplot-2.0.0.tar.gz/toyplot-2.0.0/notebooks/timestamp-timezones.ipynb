```python
import logging

import arrow
import numpy
import toyplot
```

```python
logging.basicConfig(level=logging.DEBUG)
toyplot.log.level = logging.DEBUG
```

```python
start = arrow.Arrow(2018, 7, 2, tzinfo="us/eastern")
end = arrow.Arrow(2018, 7, 7, tzinfo="us/eastern")
datetimes = list(arrow.Arrow.span_range("day", start, end))
timestamps = [datetime[0].timestamp for datetime in datetimes]
```

```python
[(arrow.get(timestamp), arrow.get(timestamp).format("dddd")) for timestamp in timestamps]
```

```python
canvas, axes, mark = toyplot.scatterplot(timestamps, numpy.zeros_like(timestamps), width=800, height=200, padding=4)
axes.x.ticks.locator = toyplot.locator.Timestamp(interval=(1, "day"), timezone="us/eastern")
axes.x.ticks.show = True
axes.x.ticks.labels.angle = 90
```
