```python
import numpy
import toyplot
```

```python
canvas = toyplot.Canvas(width=400, height=200)
numberline = canvas.numberline(label="Label")
```

```python
canvas = toyplot.Canvas(width=300)
axes = canvas.axes(xlabel="Label", ylabel="Label")
axes.plot(numpy.linspace(0, 1) ** 2)
```

```python
canvas = toyplot.Canvas(width=600, height=300)
axes = canvas.axes()
axes.plot(numpy.linspace(0, 1) ** 2, color="black")
axes = axes.share("x")
axes.y.spine.style={"stroke":"steelblue"}
axes.plot(100 - numpy.linspace(0, 10) ** 2, color="steelblue")
```
