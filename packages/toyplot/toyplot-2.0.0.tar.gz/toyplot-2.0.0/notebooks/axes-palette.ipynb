```python
import numpy
import toyplot

numpy.random.seed(1234)
```

```python
palette = toyplot.color.brewer.palette("Spectral")
palette
```

```python
canvas = toyplot.Canvas(width=600, height=150)
axes = canvas.numberline(palette=palette)
axes.scatterplot(numpy.random.normal(loc=0, size=100))
axes.scatterplot(numpy.random.normal(loc=0.5, size=100));
```

```python
canvas = toyplot.Canvas(width=600, height=300)
axes = canvas.cartesian(palette=palette)
axes.plot(numpy.cumsum(numpy.random.normal(loc=0, size=100)))
axes.plot(numpy.cumsum(numpy.random.normal(loc=0.1, size=100)));
```
