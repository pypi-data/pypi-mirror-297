```python
import numpy as np
import toyplot

palette = toyplot.color.Palette(['#9467bd', '#d62728', '#E6D812', '#2ca02c', '#1f77b4', '#ff7f0e'])
colormap = toyplot.color.CategoricalMap(palette)

ndatap = 10
color = np.random.randint(0,6, ndatap)

canvas = toyplot.Canvas(width=900, height=300)
axes = canvas.cartesian(grid=(1, 3, 0))
x = np.random.randn(ndatap)
y = np.random.randn(ndatap)
mark = axes.scatterplot(x, y, color=(color, colormap))
```
