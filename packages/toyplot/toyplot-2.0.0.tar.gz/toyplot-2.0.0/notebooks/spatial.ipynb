```python
import numpy
import scipy.spatial
import toyplot
```

```python
numpy.random.seed(1234)
coordinates = numpy.random.normal(size=(100, 2))
color = numpy.tile("black", len(coordinates)).astype("S10")

kdtree = scipy.spatial.KDTree(coordinates)
#neighbors = kdtree.query_ball_point((0, 0), r=1)
distances, neighbors = kdtree.query((0, 0), 5)
color[neighbors] = "orange"

canvas = toyplot.Canvas()
axes = canvas.axes(aspect="fit-range")
axes.scatterplot(coordinates[:,0], coordinates[:,1], color=color)
axes.scatterplot(0, 0, color="red")
```

```python
def convert_tree(node):
    if hasattr(node, "split_dim"):
        return [node.split_dim, node.split, convert_tree(node.less), convert_tree(node.greater)]
    return node.idx.tolist()

convert_tree(kdtree.tree)
```
