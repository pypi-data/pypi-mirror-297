```python
import IPython.display as display
import numpy

import toyplot
```

```python
canvas = toyplot.Canvas()
axes = canvas.cartesian()
axes.plot(numpy.arange(10));
```

```python
canvas = toyplot.Canvas(width=800, height=400)
axes = canvas.cartesian(grid=(1, 2, 0), ylabel="Axis 1")
axes = axes.share("x", ylabel="Axis 2")
axes = canvas.cartesian(grid=(1, 2, 1), ylabel="Axis 3")
```

```python
canvas = toyplot.Canvas(width=800, height=400)
axes = canvas.cartesian()
axes.scatterplot(numpy.arange(50, 100), numpy.arange(50, 100), color="crimson")
axes = axes.share("x")
axes.scatterplot(numpy.arange(10, 60), numpy.arange(50, 100), color="royalblue")
```
