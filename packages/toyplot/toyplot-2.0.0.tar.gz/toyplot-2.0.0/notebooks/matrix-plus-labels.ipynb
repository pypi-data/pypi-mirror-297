```python
import numpy
import toyplot
```

```python
numpy.random.seed(1235)
data = numpy.random.normal(size=(4, 4))
```

A matrix visualization is a table with the color of each cell mapped from the cell value, and the cell contents hidden:

```python
palette = toyplot.color.Palette(["lightblue", "white"])
colormap = toyplot.color.LinearMap(palette, domain_min=data.min(), domain_max=data.max())
```

```python
canvas, table = toyplot.matrix((data, colormap), width=400)
```

To show the cell contents, you just have to specify how to format them (they are hidden using `toyplot.format.NullFormatter`):

```python
canvas, table = toyplot.matrix((data, colormap), width=400)
table.body.cell[...].format = toyplot.format.FloatFormatter(format="{:.2g}")
```

Of course, you could choose to do this for some cells and not others, and/or override the data:

```python
canvas, table = toyplot.matrix((data, colormap), width=400)
table.body.cell[1,:].format = toyplot.format.FloatFormatter(format="{:.2g}")
table.body.cell[2,2].data = "A"
table.body.cell[2,2].format = toyplot.format.DefaultFormatter()
table.body.cell[2,2].lstyle = {"fill":"red", "font-weight":"bold", "font-size":"16px"}
```
